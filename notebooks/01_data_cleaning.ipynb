Load Data

```python
import pandas as pd

df = pd.read_csv("../data/raw data/train.csv")
df.head()
```

Convert Order Date to datetime

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)
```

```python
df = df[["Order Date", "Sales", "Category", "Region"]]
```

```python
df.head()
```

Checking missing values

```python
df.isnull().sum()
```

Create Monthly Sales Time Series

```python
# Set date as index
df.set_index("Order Date", inplace=True)

# Monthly total sales
monthly_sales = df.resample("M").sum()

monthly_sales.head()
```

Save Cleaned Data

```python
monthly_sales.to_csv("../data/processed data/monthly_sales.csv")
```
