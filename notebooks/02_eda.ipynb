Load the processed data

```python
import pandas as pd
import matplotlib.pyplot as plt

monthly = pd.read_csv("../data/processed data/monthly_sales.csv", index_col=0, parse_dates=True)

monthly.head()
```

Sales Trend Over Time

```python
monthly.plot(figsize=(12,6))
plt.title("Monthly Sales Trend")
plt.ylabel("Sales")
plt.xlabel("Date")
plt.show()
```

Seasonality (Monthly Pattern)

```python
monthly["Month"] = monthly.index.month

seasonality = monthly.groupby("Month")["Sales"].mean()

seasonality.plot(kind="bar", figsize=(10,5))
plt.title("Average Monthly Sales")
plt.ylabel("Sales")
plt.xlabel("Month")
plt.show()
```

Category-wise Sales

```python
df = pd.read_csv("../data/raw data/train.csv")

category_sales = df.groupby("Category")["Sales"].sum()
category_sales.plot(kind="bar", figsize=(8,5))
plt.title("Total Sales by Category")
plt.show()
```

Region-wise Sales

```python
region_sales = df.groupby("Region")["Sales"].sum()
region_sales.plot(kind="pie", autopct="%1.1f%%", figsize=(6,6))
plt.title("Sales by Region")
plt.show()
```
